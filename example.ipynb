```python
from functions import *
trigger_words = ['data', 'machine learning', 'ai', 'ml ', 'statist', 'artificial intelligence', 'python']
# The trigger words are used to filter job titles; For a job title to be considered, it needs to contain at least one of the trigger words.

keyword = 'Data science'
# The keyword to search for on LinkedIn

skills = ['python', 'sql', 'machine learning', 'data analysis', 'statistics', 'matplotlib', 'pandas',
    'numpy', 'scikit-learn', 'tensorflow', 'pytorch', 'keras', 'data visualization', 'data wrangling', 
    'big data', 'cloud computing', 'pyspark', 'hadoop', 'spark', 'tableau', 'power bi',
    'data mining', 'data engineering', 'data modeling', 'data governance', 'data quality',
    'data architecture', 'data strategy', 'data storytelling', 'data ethics', 'data privacy',
    'data security', 'data compliance', 'data management', 'data integration', 'data transformation',
    'data pipeline', 'data lake', 'data warehouse', 'data mart', 'data catalog', 'data lineage',
    'data profiling', 'data cleansing', 'data enrichment', 'data visualization tools',
    'business intelligence tools', 'machine learning algorithms', 'deep learning algorithms',
    'natural language processing (NLP)', 'computer vision', 'reinforcement learning']
# The skills to search for in the job descriptions (relevant in the job_info_extractor function)
# Skill can also automatically be extracted from job descriptions using keyword skills = 'extract', although this feature is experimental and not always accurate.

job_ids = get_job_ids(trigger_words, keyword, search_count=350, headers=None, internship=False, blacklist=True)
# This function retrieves job IDs from LinkedIn based on the trigger words and keyword.
# The search_count parameter controls how many jobs to fetch. The internship parameter filters out internships.
# The headers parameter is used to pass custom headers for the request, but if that doesn't tell you anything just leave it as None or 'debug'.

data = fetch_job_details(job_ids, headers=None)
# This function fetches the job details for the given job IDs.
# The headers parameter is used to pass custom headers for the request, but if that doesn't tell you anything just leave it as None or 'debug'.
# It returns a pandas DataFrame with the job details.

data_info = job_info_extractor(data, skills = skills)
# This function extracts relevant information from the job details DataFrame.
# The skills parameter is used to filter the job descriptions based on the skills provided.
# Save the data to CSV

save_jobs(data_info, keyword, date=True)
# This function saves the job details DataFrame to a CSV file.
# The keyword parameter is used to name the file, and the date parameter adds a timestamp to the filename.
# If date is set to False, the file will be saved without a timestamp, if True, the file will be saved with a timestamp.
# You can also manually imput the date in the format 'dd-mm-yyyy'.

blacklist_job_ids(job_title = 'data scientist', rows=None, cleanup=True, date=True)
# This function blacklists job IDs that have been retrieved this round and saves them to a text file.
# If blacklist is enabled in the get_job_ids function, the job IDs in the blacklist text file will be ignored in the next search.
# This is convenient to avoid duplicates in the next search.
# The rows parameter controls how many job IDs to blacklist. If set to None, all job IDs will be blacklisted.
# The cleanup parameter removes job IDs from the blacklist that are not in the current search (aka they have been removed).
# This is done to prevent the blacklist from growing indefinitely. 
# The date parameter is used to retrieve the csv correctly. If the csv was saved without a date, set to False.
# If it was saved with a date, and that date is today, set to True.
# If the date is not today, you can manually input the date in the format 'dd-mm-yyyy'.
```
